## custom function

```python
def simpleAdd(x, y):
    """
    
    a function to add 2 numbers, x and y
    """
    return (x+y)
```

```python
simpleAdd(5,6)
```

```python
def compare(x, y):
    """
    
    This is a function to compare 2 numbers, x and y
    """
    #if statement
    if x > y:
        print(f"{x} is greater than {y}")
    elif x < y:
        print(f"{x} is smaller than {y}")
    else:
        print(f"{x} is the same as {y}")
```

```python
compare(4,5)
```

## Create custom images

```python
#!pip install matplotlip
import matplotlib.pyplot as plt
```

```python
# create a black 2D image (grayscale image)
black_img = np.zeros((200,200), dtype = np.uint8)

plt.imshow(black_img, cmap=plt.cm.gray)
plt.title("black grayscale")
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# create a white image
white_img = 255 * np.ones((200,200))
white = np.uint8(white_img)

plt.imshow(white_img, cmap=plt.cm.gray, vmin = 0, vmax = 256)
plt.title("white grayscale")
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# create a random color image
random_color_image = np.random.randint(0, high = 256, size = (200, 200, 3), dtype = "uint8")

plt.imshow(random_color_image,)
plt.title("random_color")
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Exercise 1
def isOddorEven(x):
    # if x is int type
    if isinstance(x, int):
        if x == 0:
            print(f"{x} is neither even nor odd")
        elif x % 2 == 0:
            print(f"{x} is even")
        else:
            print(f"{x} is odd")
    else:
        print(f"{x} is not an integer")
```

```python
isOddorEven(0)
```

```python
isOddorEven(2)
```

```python
isOddorEven(3)
```

```python
isOddorEven(0.2)
```

```python
#Exercise 4
import matplotlib.pyplot as plt
import numpy as np

width = 200
height = 200
radius = 50

black_img = np.zeros((width,height), dtype = np.uint8)

# circle is a subclass of an Patch
circle1 = plt.Circle((height/2, width/2), radius, color='w')
# gca() -> if no Axes, a new one is created. add_patch()-> Add Patch to Axes
plt.gca().add_patch(circle1)

plt.imshow(black_img, cmap=plt.cm.gray)
plt.title("black grayscale")
plt.xticks([]), plt.yticks([])
plt.show()
```
